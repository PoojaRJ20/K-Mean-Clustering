```python
import matplotlib.pyplot as plt
x = [5,4,13,11,13,5,3,4,3,10]
y = [15,18,22,21,21,16,15,16,14,18]
plt.scatter(x,y, color='magenta')
plt.show()
```

Data Preprocessing

```python
data = list(zip(x,y))
print(data)
```

K-mean clustering

```python
!pip install scikit-learn
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, n_init='auto')
kmeans.fit(data)

c=kmeans.labels_
print(c)
```

Vizualise the clusters

```python
plt.scatter(x,y,c=kmeans.labels_)
plt.show()
```

Evalution Matrix

```python
from sklearn import metrics
metrics.silhouette_score(data, c, metric='euclidean')
```

```python
from sklearn.cluster import KMeans

data = list(zip(x,y))
inertias = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, n_init='auto')
  kmeans.fit(data)
  inertias.append(kmeans.inertia_)
print(inertias)
```

```python
plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('SSE/WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans

data = list(zip(x,y))
inertias = []

for i in range(2,10):
    kmeans = KMeans(n_clusters=i, n_init='auto')
    kmeans.fit(data)
    c=kmeans.labels_
    inertias.append(metrics.silhouette_score(data, c, metric='euclidean'))
print(inertias)
```

```python
plt.plot(range(2,10), inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('SSE/WCSS')
plt.show()
```

clustering data with higher dimensions'

```python
import matplotlib.pyplot as plt
X =[4,5,10,4,3,14,6,10,12,14]
Y = [21,19,24,17,25,24,21,22,21,20]
Z = [15,18,22,21,19,15,16,16,17,21]
from mpl_toolkits import mplot3d
ax = plt.axes(projection='3d')
ax.scatter3D(X,Y,Z, color = "green")
plt.show()
```

```python
from sklearn.cluster import KMeans

data = list(zip(X,Y,Z))
inertias = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, n_init='auto')
  kmeans.fit(data)
  inertias.append(kmeans.inertia_)
plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('inertias')
plt.show()





```

```python
from sklearn.cluster import KMeans
data = list(zip(X,Y,Z))

kmeans = KMeans(n_clusters=4, n_init='auto')
kmeans.fit(data)
c=kmeans.labels_
print(c)
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(X,Y,Z, c=kmeans.labels_)
plt.show()
```

cluster evalution with ground-truth

```python
from sklearn import metrics
metrics.silhouette_score(data, c, metric='euclidean')
```
